### Normalizing and transforming feature with MinMaxScalar() and fit_transform()

```python
# address of data 
address = '~/Desktop/2022/python-ds-essentials/2_data/mtcars.csv'
# call address
cars = pd.read_csv(address)
# name columns
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

```python
# save variable
mpg = cars.mpg

# line chart
plt.plot(mpg)
```

```python
# summary statistics 
cars[['mpg']].describe()
```

```python
# reshape 
mpg_matrix = mpg.values.reshape(-1,1)

# scale
scaled = preprocessing.MinMaxScaler()

scaled_mpg = scaled.fit_transform(mpg_matrix)

# plot
plt.plot(scaled_mpg)
```

```python
# define range 0-10
scaled = preprocessing.MinMaxScaler(feature_range=(0,10))

scaled_mpg = scaled.fit_transform(mpg_matrix)

plt.plot(scaled_mpg)
```

### Using scale() to scale your features

```python
# create variable
standardized_mpg = scale(mpg, axis=0, with_mean=False, with_std=False)

#plot
plt.plot(standardized_mpg)
```

```python
# without parameters
standardized_mpg = scale(mpg)
plt.plot(standardized_mpg)
```
