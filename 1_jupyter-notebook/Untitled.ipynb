```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
# print plots in line with jupyter notebook
%matplotlib inline

# define a figure size 8"x4"
rcParams['figure.figsize']= 8,4
```

## Labeling plot features
### The fucntional method

```python
# create variables
x = range(1,10)
y = [1,2,3,4,.5,4,3,2,1]

# generate barchart 
plt.bar(x,y)

# add labels
plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
# create variable
z = [1,2,3,4,.5]
veh_type = ['bicycle', 'motorbike', 'car', 'van', 'stroller']

# generate pie chart
plt.pie(z, labels=veh_type)

# plot pie chart
plt.show()
```

### The object-oriented method

```python
# get address data
address = '~/Desktop/2022/python-ds-essentials/2_data/mtcars.csv'

# read and save csv
cars = pd.read_csv(address)

# assign column names
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

# isolate mpg variable
mpg = cars.mpg

# create a figure
fig = plt.figure()

# add axis 
ax = fig.add_axes([.1,.1,1,1])

# call plot method
mpg.plot()

# add tick marks
ax.set_xticks(range(32))

# add lables with 60 degree rotaion
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')

# set title
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

# set x and y lables
ax.set_xlabel('car names')
ax.set_ylabel('mpg')
```

## Adding a legend to your plot 
### The functional method

```python
# create pie chart
plt.pie(z)

# create a legend located in the best location
plt.legend(veh_type, loc='best')

# show pot
plt.show()
```

### The object-oriented method

```python
# create a figure
fig = plt.figure()

# add axis 
ax = fig.add_axes([.1,.1,1,1])

# call plot method
mpg.plot()

# add tick marks
ax.set_xticks(range(32))

# add lables with 60 degree rotaion
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')

# set title
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

# set x and y lables
ax.set_xlabel('car names')
ax.set_ylabel('mpg')

# add legend
ax.legend(loc='best')
```

```python
# create a figure
fig = plt.figure()

# add axis 
ax = fig.add_axes([.1,.1,1,1])

# call plot method
mpg.plot()

# add tick marks
ax.set_xticks(range(32))

# add lables with 60 degree rotaion
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')

# set title
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

# set x and y lables
ax.set_xlabel('car names')
ax.set_ylabel('mpg')

# add legend
ax.legend(loc='best')

# set y limit
ax.set_ylim([0,45])

# create annotation at (19,33.9) with text at (21,35) of an arrow
ax.annotate('Toyota Corolla', xy=(19,33.9), xytext=(21,35), 
            arrowprops=dict(facecolor='black', shrink=0.05))
```
